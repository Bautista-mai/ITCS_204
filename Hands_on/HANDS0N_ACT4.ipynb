```python
import matplotlib.pyplot as plt
import seaborn as sns

cities = Employees['City'].dropna().str.split(',').explode().str.strip()

most_city = cities.value_counts().head(10).reset_index()
most_city.columns = ['city', 'count']

plt.figure(figsize=(15,4))
sns.barplot(data=most_city, x='city', y='count')
plt.xticks(rotation=45, ha='right')
plt.title("Top Cities That Have the Most Employees")
plt.tight_layout()
plt.show()

```

The bar graph shows that more than 2000 workers are in  Bangalore which leads by a wide margin, establishing itself as the hub for talent and operations. Pune and New Delhi, on the other hand, have significantly fewer but almost equal numbers of employees—both under 1500—indicating secondary roles in the company's regional strategy.  This concentration in Bangalore might be the result of strategic business priorities, improved infrastructure, or easier access to skilled labor.  

```python
Top_year = Employees['JoiningYear'].value_counts().nlargest(8).reset_index()
Top_year.columns = ['JoiningYear', 'Count']

plt.figure(figsize=(15, 4))
sns.barplot(data=Top_year, x='JoiningYear', y='Count')
plt.title("Top 8 Joining Years with Highest Membership")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

The bar plot shows interesting trends in the patterns of employee enrollment over time.  2017 is notable for being the year with the highest number of hires, suggesting a notable expansion effort or spike in recruitment during that time.  Consistently high joining counts from 2013 to 2015 also point to an extended period of organizational expansion or strategic recruitment of workers.  There is a noticeable drop in 2018 after the 2017 peak, which might be due to internal reorganization, a change in hiring practices, or the state of the market.  The grouping of high joining years between 2012 and 2017 might indicate intentional hiring cycles in line with project deadlines or business objectives.

```python
import pandas as pd

# CSV FILE
Employees = pd.read_csv("../Hands_on/assets_csv/Employee.csv")

Employees


plt.figure(figsize=(15,4))
plt.pie(Employees['Age'].value_counts().head(10), labels = Employees['Age'].value_counts().head(10).index, autopct='%1.1f%%')

plt.axis('equal')
plt.title("Top 10 Age Distribution of Employees")
plt.tight_layout()
plt.show()
```

The pie chart shows the top 10 age distribution of employees. This data shows that most of the employees are at the age of 26, 27, and 28. This means that most of the employees are in their mid-20's and in the of peak of their career. This age are starting to build a family and seek financial stability. 

```python
Laeve_of_Employees = Employees.groupby('Gender')['LeaveOrNot'].mean().reset_index()

sns.lineplot(data=Laeve_of_Employees, x='Gender', y='LeaveOrNot', color='green', ls="solid", marker='o')
plt.title("Employee Leave Rate by Gender")
plt.xlabel('Gender')
plt.ylabel('Leave Rate (0 - 1)')
plt.grid(axis='y', linestyle=':', alpha=0.7)

plt.gcf().set_size_inches(8, 4)
plt.show()


```

The leave rate in female is higher than male. This indicates that women are most likely to resign due to unknown factors. This data shows that there's an issue about why the leave rate of a female employees are higher more than the male. If this issue doesn't resolve this could lead into the shortage of employee, productivity, and functionality.
