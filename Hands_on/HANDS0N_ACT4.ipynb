```python
import matplotlib.pyplot as plt
import seaborn as sns


Movies_data = pd.read_csv("../Hands_on/assets_csv/Top_Rated_Movies.csv")
vote = Movies_data['title']
top_genres = vote.value_counts().head(10).reset_index()
top_genres.columns = ['title', 'vote_count']

plt.figure(figsize=(15,4))
sns.barplot(data=top_genres, x='title', y='vote_count')
plt.xticks(rotation=45, ha='right')
plt.title("MOST POPULAR MOVIE")
plt.tight_layout()
plt.show()
```

```python
Movies_data = pd.read_csv("../Hands_on/assets_csv/Top_Rated_Movies.csv")

plt.figure(figsize=(15,4))
plt.pie(Movies_data['popularity'].value_counts().head(8), labels=Movies_data['title'].value_counts().head(8).index, autopct='%1.1f%%')
plt.axis('equal')
plt.title("Top Movies")
plt.tight_layout()
plt.show()
```

```python
anime_data = Movies_data[pd.to_numeric(Movies_data["vote_average"], errors="coerce").notnull()]
anime_data["vote_average"] = Movies_data["vote_average"].astype(int)
anime_data = Movies_data[Movies_data["vote_average"] > 0]
anime_data.loc[Movies_data["vote_average"] == 0, "vote_average"] = 1

plt.Figure(figsize=(15,4))
sns.scatterplot(data=Movies_data, x='vote_count', y='vote_average')
plt.title("The difference between the vote count and the average count")
plt.xlabel(" average count")
plt.ylabel("vote_average")
plt.gcf().set_size_inches(13,8)
plt.tight_layout()
plt.show()
```
