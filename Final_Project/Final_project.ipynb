## **STATEMENT OF THE PROBLEM**



This project aims to analyze the Students Performance on Exam dataset to determine how demographic characteristics and educational support factors relate to students’ math, reading, and writing scores. The main problem of the study is Which student characteristics and contextual factors are associated with differences in academic performance across math, reading, and writing exams?

## 1. How do math, reading, and writing scores differ between male and female students?

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns

gender_scores = students_data.groupby('gender')[['math score', 'reading score', 'writing score']].mean().reset_index()

plt.figure(figsize=(15, 4))
genderscores_m = gender_scores.melt(id_vars='gender', value_vars=['math score', 'reading score', 'writing score'],var_name='Subject', value_name='Average Score')

sns.barplot(data=genderscores_m, x='Subject', y='Average Score', hue='gender')

plt.title("Average Exam Scores by Gender")
plt.ylabel("Average Score")
plt.xlabel("Subject")
plt.legend(title="Gender")
plt.tight_layout()
plt.show()
```

# INTERPRETATION

The bar plot shows the average  scores by gender, a clear pattern will stand out you will see the difference. Girls usually score higher than boys in reading and writing. Boys, on the other hand, edge out girls in math, though not by much. This isn’t just a one off result educational research keeps turning up these same trends across different studies. The gaps aren’t huge, but they stick around in subject after subject.

# CONCLUSION

In conclusion Gender shapes how students do in certain subjects. Both girls and boys do well on their subjects, but girls tend to stand out in literacy, while boys have a small edge in math(OECD(2019)). These trends point to the impact of social expectations and learning styles tied to gender.

# RECOMMENDATION

- Encourage girls’ participation in STEM through clubs, mentorship, and hands‑on math activities.
- Support boys’ literacy development through reading programs tailored to their interests.
- Promote gender‑inclusive teaching strategies that avoid reinforcing stereotypes.

Reference: - OECD. (2019). PISA 2018 Results. OECD Publishing.



## 2. Does completing a test preparation course improve students’ exam scores?

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate average scores based on test preparation course
preparation_scores = students_data.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean().reset_index()

prep_m = preparation_scores.melt(id_vars='test preparation course',value_vars=['math score', 'reading score', 'writing score'],var_name='Subject',value_name='Average Score')

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(data=prep_m, x='Subject', y='Average Score', hue='test preparation course')

plt.title("Average Exam Scores Based on Test Preparation Course Completion")
plt.xlabel("Subject")
plt.ylabel("Average Score")
plt.legend(title="Test Preparation Course")
plt.tight_layout()
plt.show()
```

# INTERPRETATION

The data shows that the students who took the test preparation course did well in their math, reading, and writing scores, especially in math. Structured preparedness doesn’t just a requirement, it gets students comfortable with the test and gives them a real shot of confidence.

# CONCLUSION

In conclusion test preparation programs are effective in terms of improving student performance across different subjects. According to College Board(2017) Students who participated in test preparation are more at advantage than to those not take it. It indicates that  access to preparation resources can significantly influence the outcomes.

# RECOMMENDATION

- Schools should expand access to free or low‑cost test preparation programs.
- Integrate test‑taking strategies into regular classroom instruction.
- Provide online practice tools for students who cannot attend in‑person sessions.

Reference: - College Board. (2017). SAT Suite of Assessments Annual Report.



## 3. How are reading and writing scores related??

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 4))

sns.scatterplot(
    data=students_data,
    x='reading score',
    y='writing score',
    hue='gender',       
    alpha=0.7
)

plt.title("Relationship Between Reading and Writing Scores")
plt.xlabel("Reading Score")
plt.ylabel("Writing Score")
plt.tight_layout()
plt.show()
```

# INTERPRETATION

The scatter plot shows that reading and writing scores have a strong positive relationship. When a student’s reading score is high, their writing score is usually high too. Both genders follow the same pattern, so the connection between the two subjects is strong for everyone.

# CONCLUSION

Therefore, the graph proves that reading and writing skills grow together. Students who struggle in one area often struggle in the other, and gender doesn’t make a big difference. The scores mostly depend on individual ability and learning support.

# RECOMMENDATION

- Teachers should combine reading and writing activities since they help each other.
- Students with low reading scores should also get writing support.
- Encourage simple daily practice like reading short texts and writing reflections.

References: Graham, S. & Hebert, M. (2011). Writing to Read.

Snow, C. (2010). Reading and Writing: The Development of Literacy.



## 4. Do students who receive free lunch score lower than those with standard lunch??

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))

lunch_m = students_data.melt(id_vars='lunch',value_vars=['math score', 'reading score', 'writing score'],var_name='Subject',value_name='Score')
sns.boxplot(data=lunch_m,x='Subject',y='Score',hue='lunch')
plt.title("Exam Scores by Lunch Status (Standard vs Free)")
plt.xlabel("Subject")
plt.ylabel("Score")
plt.legend(title="Lunch Type")
plt.tight_layout()
plt.show()
```

# INTERPRETATION

The boxplot shows that students with standard lunch usually score higher in math, reading, and writing compared to students who get free lunch. The median scores for the standard lunch group are higher in all three subjects, and their score ranges look slightly better too. Students with free lunch tend to have lower overall performance, which might be connected to economic factors since free lunch is often given to lower income students.


# CONCLUSION

In conclusion the graph suggests that lunch status (which is also related to socioeconomic status) has an impact on academic performance. Students with standard lunch generally perform better, while those with free lunch show lower scores across subjects. This means that outside factors like family income and access to resources might be affecting how well students do in school.

Recommendations
- Schools should give extra academic support to students receiving free lunch, like tutoring or after‑class help.
- Provide nutritious meals and learning resources to reduce the gap caused by economic differences.
- Teachers can monitor students who might be struggling and offer more encouragement or learning strategies.
